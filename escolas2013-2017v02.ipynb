```python
def readcsv(filename, dictFeatures, listFeatures) :
    
    with open(filename , 'r') as file :
        df = pd.read_csv(file, usecols=listFeatures,
                                  header=0, index_col=None, sep=';', 
                                  low_memory=False, encoding='utf-8', 
                                  na_filter=True)
    
    for feature in df.columns :
        if dictFeatures[feature] == int :
            df[feature] = df[feature].fillna(0).astype(int)
        if dictFeatures[feature] == float :
            df[feature] = df[feature].fillna(0).astype(float)
        if dictFeatures[feature] == bool :
            df[feature] = df[feature].astype(bool)
        if dictFeatures[feature] == "category" :
            df[feature] = df[feature].astype("category")
            
    return df  
    
```

# ADEQUAÇÃO PROFESSOR



```python
def readexcel(filename, dictFeatures) :
    
    df = pd.read_excel(filename, sheet_name=0, skiprows=[1], header=0, index_col=None)
        
    for feature in df.columns :
        if dictFeatures[feature] == int :
            df[feature] = df[feature].astype(int)
        if dictFeatures[feature] == float :
            df[feature] = df[feature].astype(float)
        if dictFeatures[feature] == bool :
            df[feature] = df[feature].astype(bool)
        if dictFeatures[feature] == "category" :
            df[feature] = df[feature].astype("category")
            
    return df 
```
